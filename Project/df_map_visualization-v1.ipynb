### pandas profiling

```python
#Importing the function
from pandas_profiling import ProfileReport

#Generate the report. We would use the mpg dataset as sample, title parameter for naming our report, and explorative parameter set to True for Deeper exploration.
starttime = time.time()

profile = ProfileReport(dfMap, title='dfMap Pandas Profiling Report', minimal=True)
# profile
print("...runtime: {:.2f} minutes".format((time.time() - starttime)/60.0))
```

```python
starttime = time.time()
profile.to_file('dfMap_profile_minimal_report.html')
print("...runtime: {:.2f} minutes".format((time.time() - starttime)/60.0))
```
