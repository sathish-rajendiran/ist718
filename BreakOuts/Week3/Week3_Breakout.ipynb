```python
np.random.seed(1234)
SalesUS['runiform'] = uniform.rvs(loc = 0, scale = 1, size = len(SalesUS))
SalesUS_train = SalesUS[SalesUS['runiform'] >= 0.33]
SalesUS_test = SalesUS[SalesUS['runiform'] < 0.33]
# check training data frame
#print('\SusItems_train data frame (rows, columns): ',susitemdf_train.shape)
#print(susitemdf_train.head())
# check test data frame
#print('\SusItems_test data frame (rows, columns): ',susitemdf_test.shape)
#print(susitemdf_test.head())

# specify a simple model with bobblehead entered last
my_model = str('SustainableClaim ~ Revenue + Quantity + SustainableMarketing')

# fit the model to the training set
train_model_fit = smf.ols(my_model, data = SalesUS_train).fit()
# summary of model fit to the training set
print(train_model_fit.summary())
```

```python
import statsmodels.formula.api as smf
model2 = smf.mixedlm("SustainableClaim ~ Revenue + Quantity + SustainableMarketing", SalesUS_train, groups=SalesUS_train["RetailerType"])

model2f = model2.fit()

print(model2f.summary())
```
